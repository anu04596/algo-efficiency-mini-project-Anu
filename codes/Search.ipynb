```python
import time
import random
import matplotlib.pyplot as plt
import numpy as np
```

Linear Search

```python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
```

Binary Search

```python
def binary_search(arr, target):
    low, high = 0, len(arr) - 1

    while low <= high:
        mid = low+ (high-low) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1
```

Input Sizes

```python
sizes = [100, 500, 1000, 5000, 10000, 20000, 50000]
linear_times = []
binary_times = []
```

Measure Time

```python
for n in sizes:
    arr = sorted(random.sample(range(1, n*10), n))
    target = arr[-1]  # worst-case: last element

    # Linear Search timing
    start = time.time()
    linear_search(arr, target)
    end = time.time()
    linear_times.append(end - start)

    # Binary Search timing
    start = time.time()
    binary_search(arr, target)
    end = time.time()
    binary_times.append(end - start)
```

Measure Space

```python
linear_space = [1 for _ in sizes]
binary_space_iterative = [1 for _ in sizes]
```

```python
plt.figure(figsize=(8,5))
plt.plot(sizes, linear_times, 'r-o', label="Linear Search")
plt.plot(sizes, binary_times, 'b-o', label="Binary Search")
plt.xlabel("Input Size (n)")
plt.ylabel("Execution Time (seconds)")
plt.title("Linear vs Binary Search Performance")
plt.legend()
plt.ylim()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(sizes, linear_space, label="Linear Search ", color="blue", marker="o")
plt.plot(sizes, binary_space_iterative, label="Binary Search Iterative", color="red", marker="o")

plt.title("Space Complexity: Linear vs Binary Search")
plt.xlabel("Input Size (n)")
plt.ylabel("Auxiliary Space Used")
plt.legend()
plt.grid(True)
plt.show()
```
