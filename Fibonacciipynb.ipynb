```python
import time
import matplotlib.pyplot as plt
```

Without Recursion

```python
def fibonacci(n):
  if n<=1 :
    return n
  a,b=0,1
  for i in range(n):
    a,b=b,a+b
  return b
```

With Recursion

```python
def fibonacci_rec(n):
  if n<=1:
    return n
  return fibonacci_rec(n-1)+fibonacci_rec(n-2)
```

Measuring Time

```python
def measure_time(func, n_values):
    times = []
    for n in n_values:
        start = time.time()
        func(n)
        end = time.time()
        times.append(end - start)
    return times
```

Input Sizes

```python
n_values_iterative = list(range(1, 100))
n_values_recursive = list(range(1, 40))
```

Measure

```python
times_iterative = measure_time(fibonacci, n_values_iterative)
times_recursive = measure_time(fibonacci_rec, n_values_recursive)
print(times_iterative)
print(times_recursive)
```

Plot

```python
plt.figure(figsize=(10,6))
plt.plot(n_values_iterative, times_iterative, label="Iterative", marker='o')
plt.plot(n_values_recursive, times_recursive, label="Recursive", marker='s', color="red")

plt.title("Fibonacci: Iterative vs Recursive Time Complexity")
plt.xlabel("n (Fibonacci term)")
plt.ylabel("Execution Time (seconds)")
plt.legend()
plt.grid(True)
plt.show()
```
